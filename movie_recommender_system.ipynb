```python

import ast
```

```python

def convert(text):
    L = []
    
    # If text is a string, convert it
    if isinstance(text, str):
        text = ast.literal_eval(text)  
    
    # Now safely iterate
    for i in text:
        if isinstance(i, dict):
            L.append(i['name'])
        else:
            L.append(i)
    
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(text):  #for print only first 3 cast
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text):  #for fetching only director name
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#convert string into list form

movies['overview'] = movies['overview'].apply(lambda x:x.split())  #split divides a string into a list
```

```python
movies.head()
```

```python
#apply transformation for genres,keyword,cast,crew
#remove spaces between words

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#new column tags concatenate the 5 columns overview, genres,keyword,cast,crew

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
```

```python
import nltk   #Natural Language Toolkit (NLTK) used for Natural Language Processing (NLP) and computational linguistics.
```

```python
from nltk.stem.porter import PorterStemmer  #porterstemmer use for its stem() method on individual words
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():  #text.split() used for convert the string to list
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)  #apply stem for tags
```

```python
#text vectorization (using bag of words technique)

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')  #max_features helps to take maximum words, stop_words helps remove the stop words(are,in etc)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()  #cv gives sparse matrix and convert it into numpy array
```

```python
#for distance between every vectors calculate cosine dist for high dimensions(5000), euclidean dist not good

from sklearn.metrics.pairwise import cosine_similarity   #0 - less similarity, 1 - high similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
#used to iterate through an iterable (like a list, tuple, or string) 
#while automatically keeping track of the index of the current item
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]  #by key=lambda x:x[1] sort by the distances not index
```

```python
#fetch the index, by index enter the similarity array then sort it and get the minimum cosine dist

def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
    
    

```
